## Text Summerization

```python
print(text.count("."))
```

```python
print(string.punctuation)
```

```python
punch = [char for char in text if char not in string.punctuation]
print(punch)
```

```python
punch = "".join(punch)
punch
```

```python
process_text = [word for word in punch.split() if word not in STOPWORDS]
process_text
```

```python
print(len(process_text))
```

```python
word_freq={}
for word in process_text:
    if word not in word_freq:
        word_freq[word]=1
    else:
        word_freq[word]+=1
word_freq
```

```python
dict(sorted(word_freq.items(),key=lambda item:item[1],reverse=True))
```

```python
max_freq = max(word_freq.values())
max_freq
```

```python
for word in word_freq.keys():
    word_freq[word]=(word_freq[word]/max_freq)
word_freq
```

```python
sent_list = nltk.sent_tokenize(text)
# sent_list
sent_freq={}
for sent in sent_list:
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_freq.keys():
            if sent not in sent_freq.keys():
                sent_freq[sent]=word_freq[word]
            else:
                sent_freq[sent]=sent_freq[sent]+word_freq[word]
sent_freq
```

```python
dict(sorted(sent_freq.items(),key=lambda item:item[1],reverse=True))
```

```python
summary_sent = nlargest(3,sent_freq,key=sent_freq.get)
summary_sent
```

```python
summary = " ".join(summary_sent)
summary
```
